## LSTM Forecasting for Junction Traffic

In this notebook, I build and evaluate an LSTM model for forecasting vehicle counts across the four different traffic junctions.

The code cell below contains a loop that trains and evaluates the LSTM model on all four junctions. The loop includes:

### Data Preprocessing

- For each junction, I extract the vehicle count time series.
- The values are scaled using MinMaxScaler to normalize the input.
- I create supervised sequences using a sliding window of 24 hours.

### LSTM Model and Training

- I use a single-layer LSTM with 50 units and ReLU activation.
- The model is compiled with the Adam optimizer and MSE loss.
- Early stopping with patience of 5 is used to avoid overfitting.
- Training is repeated for each junction independently.
- Note: I also tested tanh activation and various different patience values, but did not experience an improvement in the performance in the model.

### Forecasting and Evaluation

- After training, I generate predictions on the test set.
- Predictions are inverse-transformed to get the original scale.
- MAE and RMSE are used to evaluate each model's performance.
- We plot actual vs. predicted traffic volumes for visual comparison.





```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
import matplotlib.pyplot as plt

# Function to create sliding window sequences
def create_lstm_dataset(series, window_size=24):
  """
  Converts time-series to 24-hour spaces that LSTM can understand
  (sliding windows)
  """
  X, y = [], []
  for i in range(len(series) - window_size):
    X.append(series[i:i + window_size])
    y.append(series[i + window_size])
  return np.array(X), np.array(y)


junctions = df["Junction"].unique()
results = []

for j in junctions:

  print(f"\nTraining LSTM for Junction {j}...")

  # Select the data for the specific junction
  df_j = df[df["Junction"] == j].copy()

  # Convert the 'DateTime' column to datetime objects if it's not already
  df_j["DateTime"] = pd.to_datetime(df_j["DateTime"])

  # Set the 'DateTime' column as the index before resampling
  df_j = df_j.set_index("DateTime")[["Vehicles"]]

  # Resample to hourly, calculate mean, and interpolate missing values
  df_j = df_j.resample("h").mean().interpolate()


  # Normalize with a scaler
  scaler = MinMaxScaler()
  scaled = scaler.fit_transform(df_j)

  # Create sliding window sequences
  window_size = 24
  X, y = create_lstm_dataset(scaled, window_size)

  # Reshape for LSTM (samples, timesteps, features)
  X = X.reshape((X.shape[0], X.shape[1], 1))

  # Train/test split (80/20)
  split = int(0.8 * len(X))
  X_train, X_test = X[:split], X[split:]
  y_train, y_test = y[:split], y[split:]

  # Define the model and parameters
  model = Sequential()
  model.add(LSTM(50, activation='relu', input_shape=(X_train.shape[1], 1))) # 50 memory units, nonlinear activation
  model.add(Dense(1)) # output one value (vehicle count at the next hour)

  # Adaptive optimizer, mse
  model.compile(optimizer='adam', loss='mse')

  # Early stop to preven overtfitting by stopping training
  early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

  # Fit the model
  history = model.fit(
      X_train, y_train,
      validation_split=0.2,
      epochs=50,
      batch_size=32,
      callbacks=[early_stop],
      verbose=1
  )

  # Running predictions on the test set
  y_pred = model.predict(X_test)

  # Inverse transform the prediction to the original scale
  y_test_inv = scaler.inverse_transform(y_test.reshape(-1, 1))
  y_pred_inv = scaler.inverse_transform(y_pred)

  # Evaluation metrics
  mae = mean_absolute_error(y_test_inv, y_pred_inv)
  rmse = np.sqrt(mean_squared_error(y_test_inv, y_pred_inv))

  print(f"MAE: {mae:.2f}, RMSE: {rmse:.2f}")
  results.append({'Junction': j, 'MAE': mae, 'RMSE': rmse})

  # Plotting
  plt.figure(figsize=(10, 4))
  plt.plot(y_test_inv[:100], label='Actual')
  plt.plot(y_pred_inv[:100], label='Predicted', linestyle='--')
  plt.title(f"LSTM Forecast vs Actual – Junction {j}")
  plt.xlabel("Time Step")
  plt.ylabel("Vehicle Count")
  plt.legend()
  plt.show()

results_df = pd.DataFrame(results)
display(results_df)
```
